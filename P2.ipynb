```python
#Q2
def find_maximum(a, b, c):
    
    maximum = a if (a >= b and a >= c) else (b if (b >= a and b >= c) else c)
    return maximum


num1 = int(input())
num2 = int(input())
num3 = int(input())
result = find_maximum(num1, num2, num3)
print(f"The maximum of {num1}, {num2}, and {num3} is: {result}")
```
